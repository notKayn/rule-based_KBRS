**Rekomendasi Knowledge-based bermetode rule-based**



---



```python
from tabulate import tabulate

def recommend_smartphone_multi(preferensi):
    # Mulai dengan seluruh dataset
    result = df_filtered.copy()

    # Terapkan filter untuk setiap preferensi
    if 'baterai besar' in preferensi:
        result = battery_large
    if 'kamera bagus' in preferensi:
        result = good_rear_cam
    if 'android' in preferensi:
        result = android_gang
    if 'ios' in preferensi:
        result = ios_gang
    if 'penyimpanan besar' in preferensi:
        result = big_storage
    if 'ram besar' in preferensi:
        result = big_ram
    if 'layar mulus' in preferensi:
        result = big_hertz

    # Pastikan tidak kosong
    if result.empty:
        print("Tidak ada smartphone yang cocok dengan preferensi kamu.")
        return

    # Ambil 5 termurah
    top_5_cheapest = result.sort_values(by='Harga', ascending=True).head(5)

    # Ambil 5 termahal
    top_5_expensive = result.sort_values(by='Harga', ascending=False).head(5)

    # Gabungkan hasil
    combined_results = pd.concat([top_5_cheapest, top_5_expensive])

    # Tampilkan hasil
    print("Rekomendasi berdasarkan preferensi kamu:")
    print(tabulate(combined_results, headers='keys', tablefmt='pretty'))
```
